# 入力パラメータ

|パラメータ|型|説明|
|---|---|---|
|img_size|int|生成する画像、動画のサイズ。今回は正方形の画像を生成する|
|channel|int|チャネル数は 1ならばグレースケール、3ならば RGB、それ以外はエラーになる|
|scale|float|生成する画像の画像空間の領域：画像サイズが(20, 20)で scaleが 20の場合、横幅が[-1, 1]で縦幅が[-1, 1]となる|
|time|int|時間を CPPNの入力に含むかどうか。含む場合は整数を指定し動画が生成される。含まない場合は Noneを指定し画像が生成される|
|layers_activations|dict|CPPNのレイヤー構造。レイヤー名を keyとし、活性化関数名を valueとした辞書を入力する。keyには'input'と'output'を含む必要があり、活性化関数は['linear', 'sin', 'cos', 'tan', 'tanh', 'sigmoid']の中から選ぶ|
|hidden|int|CPPNの中間層のノード数。|

```python
img_size = 256
channel = 1
scale = 256
hidden = 8
time = None
layers_activations = {
    'input': 'sin',
    'hidden_1': 'tan',
    'hidden_2': 'tanh',
    'output': 'tanh'
}

net = MLP(
    img_size,
    channel,
    scale,
    time,
    layers_activations,
    hidden,
)

net.generate()
net.image
```
